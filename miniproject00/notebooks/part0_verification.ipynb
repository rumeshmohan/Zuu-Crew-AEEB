### Environment & API Security

```python
import os
import sys
from dotenv import load_dotenv

# Project Path Setup
project_root = os.path.abspath('..')
if project_root not in sys.path:
    sys.path.insert(0, project_root)

# Load Environment Variables
load_dotenv()

# Secure API Key Audit
providers = {
    "OpenAI": os.getenv("OPENAI_API_KEY"),
    "Google Gemini": os.getenv("GEMINI_API_KEY"),
    "Groq": os.getenv("GROQ_API_KEY"),
    "Cohere": os.getenv("COHERE_API_KEY"),
    "Mistral": os.getenv("MISTRAL_API_KEY")
}

print("🛡️  API Key Audit")
print("-" * 30)
for provider, key in providers.items():
    status = "✓ Active" if key else "✗ Missing"
    # Masking key for safety while confirming existence
    preview = f"(...{key[-8:]})" if key else ""
    print(f"{provider:15s} {status:10s} {preview}")
```
